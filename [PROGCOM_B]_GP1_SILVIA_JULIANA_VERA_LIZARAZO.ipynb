```python
import random
import IPython.display as display
import ipywidgets as widgets

# Diccionario con imágenes
imagenes = {
    "Tijera": "https://i.imgur.com/NsL6lVO.png",
    "Papel": "https://i.imgur.com/T8osSyp.png",
    "Piedra": "https://i.imgur.com/V5JD7tT.png",
    "Lagarto": "https://i.imgur.com/d27R0Tk.png",
    "Spock": "https://i.imgur.com/qwKRASI.png",
    "Pistola": "https://i.imgur.com/iODLuep.png",
    "Rayo": "https://i.imgur.com/9jlxRwJ.png"
}

# Reglas del juego
reglas = {
    "Tijera": ["Papel", "Lagarto"],
    "Papel": ["Piedra", "Spock"],
    "Piedra": ["Tijera", "Lagarto"],
    "Lagarto": ["Spock", "Papel"],
    "Spock": ["Tijera", "Piedra"],
    "Pistola": ["Piedra", "Tijera", "Lagarto"],
    "Rayo": ["Pistola", "Spock", "Tijera"]
}

# Nuevos colores
colores = {
    "Tijera": "#FF6F61",
    "Papel": "#6BBDFF",
    "Piedra": "#66BB6A",
    "Lagarto": "#FFA726",
    "Spock": "#B39DDB",
    "Pistola": "#FFD54F",
    "Rayo": "#FF4081"
}

# Emojis de resultado
emojis = {
    "ganar": "🏆🎉",
    "perder": "💀💔",
    "empate": "🔄😐"
}

# Output para limpiar la pantalla
output = widgets.Output()

# Función para jugar
def jugar(eleccion):
    with output:
        display.clear_output(wait=True)  # Borrar lo anterior

        eleccion_computadora = random.choice(list(imagenes.keys()))

        # Determinar resultado
        if eleccion == eleccion_computadora:
            resultado = f"{emojis['empate']} ¡Empate! {emojis['empate']}"
            color_resultado = "#f1c40f"
            efecto = "glow-yellow"
        elif eleccion_computadora in reglas[eleccion]:
            resultado = f"{emojis['ganar']} ¡GANASTE! {emojis['ganar']}"
            color_resultado = "#2ecc71"
            efecto = "glow-green"
        else:
            resultado = f"{emojis['perder']} ¡PERDISTE! {emojis['perder']}"
            color_resultado = "#e74c3c"
            efecto = "glow-red"

        # HTML con efecto de luz en el cuadro
        html_resultado = f"""
        <div class="{efecto}" style='
            background: {colores[eleccion]};
            padding: 30px;
            border-radius: 15px;
            text-align: center;
            color: white;
            font-size: 22px;
            font-weight: bold;
            width: 60%;
            margin: auto;
            box-shadow: 0px 5px 15px rgba(0, 0, 0, 0.3);
        '>
            <p style="font-size: 26px;">Elegiste: {eleccion} 🎭</p>
            <img src='{imagenes[eleccion]}' width='120'>

            <p style="font-size: 26px; margin-top: 20px;">Computadora eligió: {eleccion_computadora} 💻</p>
            <img src='{imagenes[eleccion_computadora]}' width='120'>

            <h2 style='color: {color_resultado}; font-size: 32px; margin-top: 20px;'>{resultado}</h2>
        </div>

        <style>
            @keyframes glow-green {{
                0% {{ box-shadow: 0px 0px 10px rgba(46, 204, 113, 0.5); }}
                50% {{ box-shadow: 0px 0px 30px rgba(46, 204, 113, 1); }}
                100% {{ box-shadow: 0px 0px 10px rgba(46, 204, 113, 0.5); }}
            }}
            @keyframes glow-red {{
                0% {{ box-shadow: 0px 0px 10px rgba(231, 76, 60, 0.5); }}
                50% {{ box-shadow: 0px 0px 30px rgba(231, 76, 60, 1); }}
                100% {{ box-shadow: 0px 0px 10px rgba(231, 76, 60, 0.5); }}
            }}
            @keyframes glow-yellow {{
                0% {{ box-shadow: 0px 0px 10px rgba(241, 196, 15, 0.5); }}
                50% {{ box-shadow: 0px 0px 30px rgba(241, 196, 15, 1); }}
                100% {{ box-shadow: 0px 0px 10px rgba(241, 196, 15, 0.5); }}
            }}

            .glow-green {{ animation: glow-green 1.5s infinite alternate; }}
            .glow-red {{ animation: glow-red 1.5s infinite alternate; }}
            .glow-yellow {{ animation: glow-yellow 1.5s infinite alternate; }}
        </style>
        """

        # Mostrar resultado con efecto
        display.display(display.HTML(html_resultado))

# Botones interactivos
botones = []
for opcion in imagenes.keys():
    boton = widgets.Button(
        description=opcion,
        layout=widgets.Layout(width='120px', height='50px')
    )
    boton.style.button_color = colores[opcion]

    # Usar lambda para evitar problemas con el cierre tardío de variables
    def crear_callback(opc):
        return lambda b: jugar(opc)

    boton.on_click(crear_callback(opcion))
    botones.append(boton)

# Mostrar botones y el output
display.display(widgets.HBox(botones))
display.display(output)
```
